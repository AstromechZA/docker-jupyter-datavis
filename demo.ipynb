```python
# An example of
# - line plot
# - multiple plots / subplots
# - line styles

import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()

plt.subplot(2, 1, 1)  # rows, cols, index
plt.title('A')
plt.xlabel('B')
plt.ylabel('C')
plt.plot(
    # arbitrary number of x, y, fmt groups
    t1, f(t1), 'bo',
    t2, f(t2), 'k',
)

plt.subplot(2, 1, 2)
plt.title('D')
plt.xlabel('E')
plt.ylabel('F')
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
```

```python
# An example of 
# - Histogram

import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
samples = mu + sigma * np.random.randn(10000)

# the histogram of the data
plt.hist(samples, bins=50, normed=0, facecolor='g', alpha=0.75)  

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of Things')
plt.grid(True)
plt.show()
```

```python
# Example of seaborn histogram

import seaborn as sns

sns.distplot(samples)  # only one line different!

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of Things')
plt.grid(True)
plt.show()
```
